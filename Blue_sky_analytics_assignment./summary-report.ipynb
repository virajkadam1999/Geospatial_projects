# **Summary of the assignment**

* **ABOUT :** **The project involved noting the changes that occured in the Air Pollution levels due to covid-19 imposed restrictions.For the project, I have monitored three compounds Nitrogen Dioxide(NO2) , Sulphur Dioxide(SO2) and Formaldehyde(HCHO), which are among the most prominent anthropogenic atmospheric pollutants.** 
* **Source :** **I have queried data from Google Earth Engine using Earth Engine python API. The data is collection of Images from Copernicus Sentinel 5P, processed to level 3.**  
* **Area of Interest :** **I have performed this study around the region of Pune,India. I have also divided the region into Urban and Rural sectors to identify the changes seperately.** 
* **Period of study :** **As most of the atmospheric compounds have a seasonal trend to them, the study is confined to period that coincides with the Summer season in Pune,India. For comparing the changes, I have used the data from March-2019 to June-2019 against March-2020 to June-2020.(Also important to mention is that April 2020 to June 2020 was the period of peak of covid-19 imposed restrictions.)**

# **Key Outcomes**

* **There was a significant reduction in levels of key atmospheric pollutants such as NO2 and SO2 over the Urban as well as Rural parts of Pune during the time of covid-19 imposed restrictions.**
* **This can be attributed to almost negligible movement of people  (hence vehicles) due to lockdown restrictions imposed by the government to restrict the spread of disease.**
* **Most of the industries classified as non-essential were also not operational during the lockdown period, which would have lowered the Emissions Further.Service and other industries which can function remotely allowing work from home have also helped reduce emission by reduction in transport and office electricity consumptions.**
* **All these factors resulted in a good air quality during the period of lockdown restrictions. The emissions however have gone up again with the opening up of economy and lesser restrictions on movement of people.**
* **One more finding is that for the compounds in study, the concentration follows a seasonal trend, where the maximum concentration is reached in the winter season around the months of December/January, and the values recceed during summer months and reach a minimum in the months that coincide with the Monsoon, after which they rise again in the following winter months.**

# Nitrogen Dioxide
**Nitrogen dioxide enters the atmosphere as a result of anthropogenic activities such as fossil fuel combustion and biomass burning, as well as natural processes including microbiological processes in soils, wildfires and lightning.**

* **Link to NO2 notebook :** [ https://www.kaggle.com/virajkadam/change-in-no2-levels-during-covid-19-lockdown ] 
* **URBAN :** **Percent drop in Mean Urban NO2 level during pandemic restrictions is 27.5 %**
* **Rural :** **Percent drop in Mean Rural NO2 level during pandemic restrictions is 19 %**

# Sulphur Dioxide
**The largest source of SO2 in the atmosphere is the burning of fossil fuels by power plants and other industrial facilities. Smaller sources of SO2 emissions include: industrial processes such as extracting metal from ore; natural sources such as volcanoes; and locomotives, ships and other vehicles and heavy equipment that burn fuel with a high sulfur content.**
* **Link to SO2 Notebook :** [ https://www.kaggle.com/virajkadam/change-in-so2-levels-during-covid-19-lockdown ]
* **URBAN : Percent drop in Mean Urban SO2 level during pandemic restrictions is 23 %**
* **Rural : Percent drop in Mean Rural SO2 level during pandemic restrictions is 61 %**
